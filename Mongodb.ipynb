```python
#Get information from movies with Will Smith
protoDf = {'id' : [], 'title' : [], 'cast' : [], 'directors' : []}

col = db.movies.find()

for item in col:
    #We are telling python to retrieve the value from the 'cast' field.
    #If no field 'cast' is found, return an empty list
    cast = item.get('cast', [])
    
    if 'Will Smith' in cast:
        print(item['cast'])
        protoDf['id'].append( item['_id'])
        protoDf['title'].append( item['title'] )
        protoDf['cast'].append( item['cast'] )
        protoDf['directors'].append( item['directors'] )    
```
